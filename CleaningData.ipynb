```python
import pandas as pd
import re

# Load CSV
df = pd.read_csv('EnronSpamData.csv')

def clean_text(text):
    if not isinstance(text, str): return ""
    text = text.lower()
    text = re.sub(r'http\S+|www\S+', '', text) # Remove URL
    text = re.sub(r'[^a-z\s]', '', text) # Remove Special Characters
    return " ".join(text.split()) # Remove Extra White Space

df['subject'] = df['Subject'].fillna('no subject').apply(clean_text) #New Subject Column
df['body'] = df['Message'].fillna('').apply(clean_text) #New Body Column
df = df.drop(columns = ['Subject', 'Message']) #Removes the Old Subject and Message Column

# Convert labels: ham = 0, spam = 1
df['label'] = df['Spam/Ham'].map({'ham': 0, 'spam': 1})

df.to_csv('EnronSpamDataCleaned.csv', index=False)
```
