```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

```python
# Model performance metrics
data = {
"Model": ["SVM", "KNN", "CNN"],
"Accuracy": [0.9777, 0.6740, 0.9888],
"Precision": [0.98, 0.80, 0.99],
"Recall": [0.98, 0.67, 0.99],
"F1-Score": [0.98, 0.63, 0.99]
}

df = pd.DataFrame(data)
df
```

```python
# Reshaping data for plotting
df_melted = df.melt(id_vars="Model",
                    var_name="Metric",
                    value_name="Score")

# Creating the bar plot
plt.figure(figsize=(12, 6))
sns.barplot(data=df_melted, x="Model", y="Score", hue="Metric")

plt.title("Model Performance Comparison", fontsize=16)
plt.ylabel("Score")
plt.ylim(0.6, 0.99)
plt.legend(title="Metric")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = (8, 5))
sns.barplot(x="Model", y="Accuracy", data=df, palette="viridis")

plt.title("Accuracy Comparison of Models", fontsize=14)
plt.ylim(0.6, 0.99)

for index, value in enumerate(df["Accuracy"]):
    plt.text(index, value + 0.01, f"{value:.3f}", ha='center')

plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=df_melted, x="Model", y="Score", hue="Metric")
plt.title("Model Performance Comparison")
plt.ylim(0.6, 0.99)
plt.tight_layout()
plt.savefig("../outputs/model_comparison.png", dpi=300)
plt.show()
```
